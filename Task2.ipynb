```python
# FINANCE INVESTMENT RECOMMENDATION SYSTEM

import pandas as pd

# Creating Dataset inside code
data = {
    "Investment": [
        "Savings Account", "Fixed Deposit", "Government Bonds",
        "PPF", "Debt Mutual Fund", "Balanced Mutual Fund",
        "Index Fund", "Large Cap Stocks", "Mid Cap Stocks",
        "Small Cap Stocks", "Cryptocurrency", "Gold ETF",
        "Real Estate"
    ],

    "Category": [
        "Fixed Income", "Fixed Income", "Debt",
        "Tax Saving", "Mutual Fund", "Mutual Fund",
        "Equity", "Equity", "Equity",
        "Equity", "Digital Asset", "Commodity",
        "Property"
    ],

    "Risk_Level": [
        "Low", "Low", "Low",
        "Low", "Medium", "Medium",
        "Medium", "High", "High",
        "Very High", "Very High", "Medium",
        "High"
    ],

    "Minimum_Amount": [
        1000, 5000, 10000,
        500, 2000, 5000,
        3000, 10000, 15000,
        20000, 1000, 3000,
        500000
    ],

    "Expected_Return_%": [
        3, 6, 7,
        7.5, 9, 11,
        12, 14, 16,
        18, 25, 10,
        12
    ],

    "Liquidity": [
        "High", "Medium", "Low",
        "Low", "Medium", "Medium",
        "Medium", "Low", "Low",
        "Low", "High", "High",
        "Very Low"
    ],

    "Time_Horizon_Years": [
        0, 1, 5,
        15, 3, 5,
        7, 7, 8,
        10, 2, 4,
        10
    ]
}

# Convert to DataFrame
df = pd.DataFrame(data)

print("\nðŸ’° Welcome to Smart Investment Advisor\n")

# User Inputs
budget = int(input("Enter your investment budget (â‚¹): "))
risk = input("Enter your risk tolerance (Low / Medium / High / Very High): ").title()
time = int(input("Enter investment time horizon (years): "))

# Filtering Logic
recommend = df[
    (df["Minimum_Amount"] <= budget) &
    (df["Risk_Level"] == risk) &
    (df["Time_Horizon_Years"] <= time)
]

# Sorting by highest expected return
recommend = recommend.sort_values(by="Expected_Return_%", ascending=False)

# Output
print("\nðŸ“Š Recommended Investment Options:\n")

if not recommend.empty:
    table = recommend[["Investment", "Expected_Return_%", "Liquidity", "Time_Horizon_Years"]]
    print(table.to_string(index=False))
else:
    print("No exact match found. Try adjusting your budget or risk preference.")
```
