```python
# --- Rule-based Chatbot ---

from datetime import datetime
import re
import random

print ("Hello! I am your Chatbot")
print ("Type 'bye' to exit. \n")

user_name = " "

while True:
    user = input("You: ").lower().strip()

    # Greetings
    if user in ["hi", "hello", "hey"]:
        print("Bot: Hello! How can I help you today?")

    # Asking bot name
    elif "your name" in user:
        print("Bot: I am a rule-based chatbot created using Python.")

    elif "what is my name" in user:
        if user_name:
            print(f"Bot: Your name is {user_name.title()}.")
        else:
            print("Bot: I don't know your name yet. Tell me using 'My name is ___'.")

    # How are you
    elif "how are you" in user:
        print("Bot: I'm functioning perfectly! ðŸ¤–")

    # Date
    elif "date" in user:
        today = datetime.now().strftime("%d-%m-%Y")
        print("Bot: Today's date is", today)

    # Time
    elif "time" in user:
        now = datetime.now().strftime("%H:%M:%S")
        print("Bot: Current time is", now)

    # Simple Calculator
    elif "add" in user:
        numbers = re.findall(r'\d+', user)
        if len(numbers) >= 2:
            result = int(numbers[0]) + int(numbers[1])
            print("Bot: The sum is", result)
        else:
            print("Bot: Please provide two numbers to add.")

    # Capital of India
    elif "capital of india" in user:
        print("Bot: The capital of India is New Delhi.")

    # Sports
    elif "sports" in user:
        print("Bot: I love sports! Cricket and football are very popular in India.")

    # Tell Joke
    elif "joke" in user:
        jokes = [
            "Why did the computer go to the doctor? Because it caught a virus!",
            "Why do programmers prefer dark mode? Because light attracts bugs!",
            "Why was the social science book sad? Because it had too many theory!"
        ]
        print("Bot:", random.choice(jokes))

    # What can you do
    elif "what can you do" in user:
        print("Bot: I am a rule-based chatbot. I reply based on predefined conditions.")

    # Thank you
    elif "thank" in user:
        print("Bot: You're welcome!")

    # Exit
    elif user == "bye":
        print("Bot: Goodbye! Have a wonderful day.")
        break

    # Default
    else:
        print("Bot: Sorry, I didn't understand, Try something else.")

```
